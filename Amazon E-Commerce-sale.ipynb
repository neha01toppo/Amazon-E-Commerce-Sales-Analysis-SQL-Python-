## Data Cleaning

### Steps performed:
#### Cleaned column name 
#### Handled missing value 
#### Standardized categorical values


```python
df.columns=df.columns.str.strip().str.lower().str.replace(" ","_").str.replace("-","_").str.replace(":","_")
df.columns
```

```python
df.isnull().sum()
```

```python
df['currency'].unique()
```

```python
for col in['ship_city','ship_state','courier_status','category','status']:
 print("\n",col)
 print(df[col].value_counts())
```

```python
df['ship_city']=df['ship_city'].str.strip().str.upper()
df['ship_state']=df['ship_state'].str.strip().str.upper()
df['category']=df['category'].str.strip().str.upper()
```

```python
df['ship_city'].value_counts()
```

```python
df['ship_state'].value_counts()
```

```python
state_fix={'NEW DELHI':'DELHI',
          'ORISSA':'ODISHA',
          'RAJSHTHAN':'RAJASTHAN',
           'RAJSTHAN':'RAJASTHAN',
 'RJ':'RAJASTHAN',
 'NL':'NAGALAND',
 'PB':'PUNJAB',
 'PUNJAB/MOHALI/ZIRAKPUR':'PUNJAB',
 'APO':'ANDHRA PRADESH',
 'PONDICHERRY':'PUDUCHERRY',
 'AR':'ARUNACHAL PRADESH'
}
```

```python
df['ship_state']=df['ship_state'].replace(state_fix)
```

```python
df['ship_state'].value_counts()
```

```python
df.isnull().mean().sort_values(ascending=False)*100
```

```python
df[df['amount'].isna()]
```

```python
df.drop(columns=['unnamed__22'],inplace=True)
```
